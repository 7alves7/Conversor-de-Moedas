```python
import requests
import ipywidgets as widgets
from IPython.display import display, clear_output

# Configurações de Estilo
style = {'description_width': 'initial'}
layout = widgets.Layout(width='300px', margin='5px')

# Lista de moedas comuns
moedas = [
    ('Dólar Americano (USD)', 'USD'),
    ('Real Brasileiro (BRL)', 'BRL'),
    ('Euro (EUR)', 'EUR'),
    ('Libra Esterlina (GBP)', 'GBP'),
    ('Peso Argentino (ARS)', 'ARS'),
    ('Iene Japonês (JPY)', 'JPY'),
    ('Bitcoin (BTC)', 'BTC')
]

# Componentes da Interface
titulo = widgets.HTML("<h2>💱 Conversor de Moedas Real-time</h2>")
input_valor = widgets.FloatText(value=1.0, description="Valor:", style=style, layout=layout)
dropdown_de = widgets.Dropdown(options=moedas, value='USD', description="De:", style=style, layout=layout)
dropdown_para = widgets.Dropdown(options=moedas, value='BRL', description="Para:", style=style, layout=layout)
botao_converter = widgets.Button(description="Converter", button_style='success', layout=widgets.Layout(width='300px', height='40px', margin='10px 0px'))
output_resultado = widgets.Output()

def converter_moeda(b):
    with output_resultado:
        clear_output()
        try:
            # URL da API (usando o par de moedas selecionado)
            url = f"https://open.er-api.com/v6/latest/{dropdown_de.value}"
            response = requests.get(url)
            dados = response.json()

            if dados['result'] == 'success':
                taxa = dados['rates'][dropdown_para.value]
                resultado = input_valor.value * taxa

                # Exibição Elegante
                print(f"✅ Conversão Concluída!")
                print(f"--------------------------")
                print(f"{input_valor.value:.2f} {dropdown_de.value} = {resultado:.2f} {dropdown_para.value}")
                print(f"Taxa atual: {taxa:.4f}")
            else:
                print("❌ Erro ao buscar taxas. Tente novamente.")
        except Exception as e:
            print(f"⚠️ Erro de conexão: {e}")

# Vincular evento ao botão
botao_converter.on_click(converter_moeda)

# Organizar e exibir a interface
interface = widgets.VBox([
    titulo,
    input_valor,
    dropdown_de,
    dropdown_para,
    botao_converter,
    output_resultado
], layout=widgets.Layout(align_items='center', border='1px solid #ddd', padding='20px', width='400px'))

display(interface)
```
